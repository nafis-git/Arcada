### Object Oriented Interface
Every figure is divided into some objects and the object hierarchy is clear.

**Figure in matplotlib is divided into two different objects.**


1.   Figure object :A Figure object can contain one or more axes objects
2.   Axes object: One axes represents one plot inside figure

### The Axes Object
One or more axes objects are required to get us started with plotting
There are more then one ways an axes object can be obtained. We will start with the **add_subplot()** method

**add_subplot(num_rows, num_cols, subplot_location)** method creates a grid of of subplots of size (num_rows x num_cols) and returns an **axes** object for the subplot at subplot_location.
**The subplots are numbered in following way:**
- First subplot is at (first row, first column) location. Start from this position and continue numbering till the last column of the first row
- Start from the left most position on the second row and continue numbering

**Ex:** The 3rd subplot in a grid of 2x2 subplots is at location = (2nd row, 1st column)


```python
ax = fig.add_subplot(1,1,1)
# Set the title of plot
ax.set_title("Empty plot")
plt.show()
```

### Q:
We divide the figure in a 2x2 grid of subplots and get the axes object for all the subplots.

```python
import matplotlib.pyplot as plt

fig = plt.figure()

# Generate a grid of 2x2 subplots and get
# axes object for 1st location
ax1 = fig.add_subplot(2,2,1)
ax1.set_title('First Location')

# Get the axes object for subplot at 2nd
# location
ax2 = fig.add_subplot(2,2,2)
ax2.set_title('Second Location')

# Get the axes object for subplot at 3rd
# location
ax3 = fig.add_subplot(2,2,3)
ax3.set_xlabel('Third Location')

# Get the axes object for subplot at 4th
# location
ax4 = fig.add_subplot(2,2,4)
ax4.set_xlabel('Fourth Location')

# Display
plt.show()
```

**Some of methods of the axes objects** https://matplotlib.org/stable/api/axes_api.html
- plot(x, y) : Generate y vs x graph-
- set_xlabel() : Label for the X-axis
- set_ylabel() : Label for the Y-axis
- set_title() : Title of the plot
- legend() : Generate legend for the graph
- hist() : Generate histogram plot
- scatter(): Generate scatter plot

## Ex1 :
A Simple XY Plot

```python
import matplotlib.pyplot as plt

# Generate data for plots
x = [1, 2, 3, 4, 5]
y = x

# Get an empty figure
fig1 = plt.figure()

# Get the axes instance at 1st location in 1x1 grid
ax = fig1.add_subplot(1,1,1)

# Generate the plot
ax.plot(x, y)

# Set labels for x and y axis
ax.set_xlabel('X--->')
ax.set_ylabel('Y--->')

# Set title for the plot
ax.set_title('Simple XY plot')

# Display the figure
plt.show()
```

```python
# Generate data for plots
import numpy as np

x = np.array([1, 2, 3, 4, 5])
y = x
z = x**2
# Multiple subplots in same figure
fig3 = plt.figure()
# Divide the figure into 1 row 2 column grid and get the
# axes object for the first column
ax1 = fig3.add_subplot(1,2,1)

# plot y = x on axes instance 1
ax1.plot(x, y)

# set x and y axis labels
ax1.set_xlabel('X------>')
ax1.set_ylabel('Y------>')
ax1.set_title('y=x plot')

# Get second axes instance in the second column of the 1x2 grid
ax2 = fig3.add_subplot(1,2,2)

# plot z = x^2
ax2.plot(x, z)
ax2.set_xlabel('X---------->')
ax2.set_ylabel('z=$X^2$--------->')
ax2.set_title('$z=x^2$ plot')

# Generate the title for the Figure. Note that this is different then the title for individual plots
plt.suptitle("Two plots in a figure")
plt.show()
```

### Ex3: Histogram Plots
Data for this example is generated using numpy. 1000 samples are generated from a gaussian distribution with mean of 10 and standard deviation of 0.5.

```python
# coding: utf-8
import matplotlib.pyplot as plt
import numpy as np


# Generate 1000 numbers from gaussian sample
mean = 10
std = 0.5
num_samples = 1000
samples = np.random.normal(mean, std, num_samples)

# Get an instance of Figure object
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

# Generate histogram plot
ax.hist(samples)
ax.set_xlabel('Sample values')
ax.set_ylabel('Frequency')
ax.set_title('Histogram plot')

plt.show()
```

```python
# or we can write a function for further use
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
  axes.plot(x, y, color=color)
  axes.set_xlabel(xlabel)
  axes.set_ylabel(ylabel, color=color)
  axes.tick_params('y', colors=color)
```

### Q:
Are there rowers with a height of less than 180 centimeters? show it with histogram

```python
# plotting histogram
fig, ax = plt.subplots()
# histtype of "step", which displays the histogram as thin lines, instead of solid bars,
ax.hist(mens_gym['Height'], label = 'Gymnastic', bins = [150,160,170,180,190,200,210],histtype = 'step' ) # default of bins is 10
ax.hist(mens_rowing['Height'], label = 'Rowing', bins = [150,160,170,180,190,200,210], histtype = 'step' )
ax.set_xlabel('height (cm)')
ax.set_ylabel('# of observation')
ax.legend()
```

# histogram interpretation
The x-axis is the values within the variable and the height of the bars represents the number of observations within a particular bin of values. For example, there are 12 gymnasts with heights between 164 and 167 centimeters, so the highest bar in the orange histogram is 12 units high. Similarly, there are 20 rowers with heights between 188 and 192 centimeters, and the highest bar in the blue histogram is 20 units high.

### Adding error bars to bar charts
There are at least two different ways to display error bars.


1.   add the error bar as an argument to a bar chart (yerr= )
2.   Error bars in plots (ax.errorbar(x,y,yerr))



```python
# plot a barplot with an error of its standard deviation
fig, ax = plt.subplots()
ax.bar("Rowing",  mens_rowing["Height"].mean(),
       yerr=mens_rowing["Height"].std())
ax.bar("Gymnastics", mens_gym["Height"].mean(),
       yerr=mens_gym["Height"].std())
ax.set_ylabel("Height (cm)")
plt.show()
```

```python
# plot a plot with errorbar
fig, ax = plt.subplots()
ax.errorbar(seattle_weather["DATE"],
            seattle_weather["MLY-TAVG-NORMAL"],
            yerr=seattle_weather["MLY-TAVG-STDDEV"])
ax.errorbar(austin_weather["DATE"],
            austin_weather["MLY-TAVG-NORMAL"],
            yerr=austin_weather["MLY-TAVG-STDDEV"])
ax.set_ylabel("Temperature (Fahrenheit)")
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample dataframe
import pandas as pd
import numpy as np

np.random.seed(0)
data = pd.DataFrame({'x': np.random.randn(100),
                     'y': np.random.randn(100),
                     'z1': np.random.rand(100),
                     'z2': np.random.rand(100)})

# Create two separate plots
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Plot for the first axis
sns.scatterplot(ax=axes[0], x="x", y="y", hue="z1", data=data)
axes[0].set_title("Relational Plot 1")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Y")
axes[0].legend(title="z1")

# Add color bar for the first axis
sm = plt.cm.ScalarMappable(cmap="viridis", norm=plt.Normalize(vmin=data['z1'].min(), vmax=data['z1'].max()))
sm.set_array([])
fig.colorbar(sm, ax=axes[0], label='z1')

# Plot for the second axis
sns.scatterplot(ax=axes[1], x="x", y="y", hue="z2", data=data)
axes[1].set_title("Relational Plot 2")
axes[1].set_xlabel("X")
axes[1].set_ylabel("Y")
axes[1].legend(title="z2")

# Add color bar for the second axis
sm = plt.cm.ScalarMappable(cmap="viridis", norm=plt.Normalize(vmin=data['z2'].min(), vmax=data['z2'].max()))
sm.set_array([])
fig.colorbar(sm, ax=axes[1], label='z2')

plt.tight_layout()
plt.show()
```

Point plots vs. lineplots
Both show:
> 95% condence intervals for the mean
> Mean of quantitative variable

 Differences:
 > Lineplot has quantitative variabl (usuallytime) on x-axis
 > Pointplot has categorical variable on x-axis


```python
# Create a point plot of family relationship vs. absences
sns.catplot(x = "famrel", y = "absences" , data = student_data, kind ='point' )

# Show plot
plt.show()
```

```python
# Add "caps" to the end of the confidence intervals with size 0.2.
# Add caps to the confidence interval
sns.catplot(x="famrel", y="absences",
			data=student_data,
            kind="point", capsize = 0.2)

# Show plot
plt.show()
```

```python
#Use sns.catplot() and the student_data DataFrame to create a point plot with relationship status ("romantic") on the x-axis and number of absences ("absences") on the y-axis. Color the points based on the school that they attend ("school").
#Since there may be outliers of students with many absences, use the median function that we've imported from numpy to display the median number of absences instead of the average.
# Import median function from numpy
from numpy import median

# Plot the median number of absences instead of the mean
sns.catplot(x="romantic", y="absences",
			data=student_data,
            kind="point",
            hue="school",
            estimator = median, ci = None)

# Show plot
plt.show()
```
